```python
def sum_number(*args):
  sum = 0
  for num in args:
    sum+=num
  return sum
```

```python
sum_number(1, 2, 3, 4, 5, 6)
```

```python
# keyword variable length argument
def test(**kwargs):
  print(kwargs)
  print(len(kwargs))
  print(type(kwargs))
```

```python
test(a = 10, b = 20, c = 30)
```

```python
def pass_student_list(**kwargs):
  pass_student_list = []
  for name, marks in kwargs.items():
    if marks>35:
      pass_student_list.append(name)
  return pass_student_list
```

```python
pass_student_list(A = 10, B = 20, C = 40, D = 50, E = 50, F=  30)
```

```python
import numpy as np
from numpy.linalg import norm
A = np.array([2, 1, 2, 3, 2, 9])
B = np.array([3, 4, 2, 4, 5, 5])

# compute cosine similarity
cosine = np.dot(A,B)/(norm(A)*norm(B))
print("cosine Similarity:",cosine)
```
